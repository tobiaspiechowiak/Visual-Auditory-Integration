```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
import matplotlib as mpl
```

```python
#data import 
path = 'data/NH/'
data_NH = pd.read_csv(path + 'NH.csv',sep = ',',header= None).dropna().values

path = 'data/DR/'
data_DR = pd.read_csv(path + 'DR.csv',sep = ',',header= None).dropna().values
```

```python
#GMM for normal hearing and plotting the results
for idx in range(1,17):
    X = data_NH[data_NH[:,2] == idx]; X = X[:,0:2] 
    bic = []
    lowest_bic = np.infty
    n_components_range = range(1,4)
    for n_comp in n_components_range:
        #fit gaussian mixture with EM
        gmm = GaussianMixture(n_components= n_comp, covariance_type= 'full')
        gmm.fit(X)
        bic.append(gmm.bic(X))
        if bic[-1] < lowest_bic:
            lowest_bic = bic[-1]
            best_gmm = gmm
        
    bic = np.array(bic)

    clf = best_gmm
    #bars = []

# Plot the BIC scores
#plt.figure(figsize=(8, 6))

#xpos = np.array(n_components_range) +  (i - 1)
#plt.bar(xpos, bic,width=.7)
#plt.ylim([bic.min() * 1.01 - .01 * bic.max(), bic.max()])
#plt.title('BIC scores subject 1')
#plot the scatter data and distribution
    plt.subplot(4,4,idx)
    Y_ = clf.predict(X)
    for i, (mean, cov) in enumerate(zip(clf.means_, clf.covariances_)):
    
        if not np.any(Y_ == i):
            continue      
        plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1], 2)
    
    plt.xticks(())
    plt.yticks(())
    plt.title('Subject ' + str(idx),fontsize=5, y=-.03,x=.75)

plt.savefig('GMM_on_NH.png',dpi=300)
```
